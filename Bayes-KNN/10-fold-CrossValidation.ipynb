### step 1: <mark>Data Wrangling</mark>

Get data from API 
and shown as a dataframe ðŸš€ 

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

iris = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
iris
```

### step 2: <mark>Data Cleaning</mark>

Save the original dataset before cleaning.

```python
iris_clean = iris.copy()
```

```python
iris_clean.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'label']
iris_clean
```

Drop label for X.

```python
X_iris = iris_clean.drop('label', axis=1)
X_iris
```

Transform label from text to number.

```python
le = LabelEncoder()
iris_clean['class'] = le.fit_transform(iris_clean['label'])
y_iris = iris_clean['class']
y_iris
```

### step 3: <mark>Create Model with NaÃ¯ve Bayes and kNN</mark>

- NaÃ¯ve Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, cross_val_score
from sklearn.naive_bayes import GaussianNB

fold = 10
iterations = 30
sum_score = 0.0

def get_dataset(seed):
    return train_test_split(X_iris, y_iris,random_state=seed) # default: 75 / 25 %

def get_avg_cross_val_score(model):
    cv = KFold(n_splits=fold, shuffle=True)
    scores = cross_val_score(model, X_iris, y_iris, cv=cv)

    score_sum = 0.0
    for i in scores:
        score_sum += i

    return score_sum / fold

def NaiveBayes():
    scores = []
    for i in range(iterations):
        Xtrain, Xtest, ytrain, ytest = get_dataset(seed=i) 
        model = GaussianNB()
        model.fit(Xtrain, ytrain)
        y_model = model.predict(Xtest)
        score = get_avg_cross_val_score(model)
        scores.append(score)
    return scores

scores = NaiveBayes()

for y in scores:
    sum_score += y;

avg_score = sum_score / iterations
print("10-fold Croos-Validation AVG score (Bayes):", avg_score)
```

- kNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, cross_val_score

fold = 10
iterations = 30
sum_score = 0.0


def get_dataset(seed):
    return train_test_split(X_iris, y_iris,random_state=seed) # default: 75 / 25 %

def get_avg_cross_val_score(model):
    cv = KFold(n_splits=fold, shuffle=True)
    scores = cross_val_score(model, X_iris, y_iris, cv=cv)

    score_sum = 0.0
    for i in scores:
        score_sum += i

    return score_sum / fold

def NaiveBayes():
    scores = []
    for i in range(iterations):
        Xtrain, Xtest, ytrain, ytest = get_dataset(seed=i) 
        model = KNeighborsClassifier(n_neighbors=1)
        model.fit(Xtrain, ytrain)
        y_model = model.predict(Xtest)
        score = get_avg_cross_val_score(model)
        scores.append(score)
    return scores

scores = NaiveBayes()

for y in scores:
    sum_score += y;

avg_score = sum_score / iterations
print("10-fold Croos-Validation AVG score (kNN):", avg_score)
```
